```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from math import cos, sin, floor
```

```python
img = cv.imread("grayscale_img.png", cv.IMREAD_GRAYSCALE)
imgOr = cv.imread("rgb_img.png")
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
imgsplit = cv.split(img)
histo = cv.calcHist(imgsplit,[0],None,[255],(0,256))
plt.plot(histo)
```
