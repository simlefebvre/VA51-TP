```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.cvtColor(cv.imread("rgb_img.png"), cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
imgGray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(imgGray, cmap='gray')
```

```python
(R,G,B) = cv.split(img)
plt.imshow(B,cmap='gray')
```
